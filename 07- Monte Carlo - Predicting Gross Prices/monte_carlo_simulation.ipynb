## Monte Carlo - Predicting Gross Profits

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Revenue in Millions:
rev_m = 170
# Standard Deviation of revenue:
rev_stdev = 20 
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
# Plot to see the observations:
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show

#centered on the mean of 170
```

```python
# Cost of Good Sold (COGS) -> approximately 60% of revenues:
# COGS are money spent so its negative
# S.D. = 0.1, COGS = 0.6

COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
print(f"COGS Mean: {COGS.mean()}")
print(f"COGS Std. Deviation: {COGS.std()}")
```

```python
#Gross Profit
# COGS is a negative number so use '+'

gross_profit = rev + COGS
gross_profit

plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()

#Normally distributed
```

```python
print(f"Biggest Potential value of Gross Profit: ${(max(gross_profit)):.3f}")
print(f"Smallest Potential value of Gross Profit: ${(min(gross_profit)):.3f}")
```

```python
print(f"Average Gross Profit: {(gross_profit.mean()):.4f}")
print(f"Std. Deviation of Gross Profit: {(gross_profit.std()):.4f}")
```

```python
#Histogram
plt.figure(figsize=(10,6))
plt.hist(gross_profit, bins = 20)
plt.show()
```

```python
# Histogram shows that it is normally distributed.
```
